```python
from nokai import NokaiConfig, NokaiModel
import torch

# Créer un modèle nano pour test
config = NokaiConfig.nano()
config.device = "cuda" if torch.cuda.is_available() else "cpu"
model = NokaiModel(config)
model.eval()

# Test avec des tokens aléatoires
input_ids = torch.randint(0, 1000, (1, 10), device=config.device)
with torch.no_grad():
    output = model.generate(input_ids, max_new_tokens=20)
print(f"Generated shape: {output.shape}")
```
